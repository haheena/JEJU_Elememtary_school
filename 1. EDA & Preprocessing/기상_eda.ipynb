```python
group_amount2 = final.groupby("행정동").mean("강수량(mm)")
```

```python
plt.figure(figsize=(10, 6))
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.plot(group_amount2["강수량(mm)"], label='월평균강수량')
plt.xticks(rotation=45)
```

```python
rain_top5 = group_amount2.sort_values(by="강수량(mm)", ascending=False).head(5)
rain_bottom5 = group_amount2.sort_values(by="강수량(mm)", ascending=False).tail(5)
```

```python
rain_top5
```

```python
rain_bottom5
```

```python
plt.figure(figsize=(10, 6))
plt.rcParams['font.family'] = 'Malgun Gothic'

plt.subplot(2, 1, 1) 
plt.plot(rain_top5["사고 발생 건수"])
plt.title('강수량 top 5 연평균사고발생건수')

plt.subplot(2, 1, 2) 
plt.plot(rain_bottom5["사고 발생 건수"])
plt.title('강수량 bottom 5 연평균사고발생건수')

plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.5)
```
