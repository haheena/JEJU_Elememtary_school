# K means clustering + pca

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 데이터 불러오기
#data = pd.read_csv('data.csv')

# 선택한 feature만 추출하여 X에 저장
selected_features = ['규제표지', '보조표지', '주의표지', '지시표지', '차로수', '신호등개수', '도로폭', '보차분리비율', '안전도지수']
X = test_df_s[selected_features]

# PCA를 사용하여 2차원으로 차원 축소
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_scaled)

# KMeans 클러스터링
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_2d)

# 클러스터링 결과 시각화
labels = kmeans.labels_
colors = ['r', 'g', 'b']
for i in range(len(colors)):
    plt.scatter(X_2d[labels == i, 0], X_2d[labels == i, 1], s=50, c=colors[i], label=f'Cluster {i+1}')

plt.title('PCA + KMeans Clustering')
plt.legend()
plt.show()
```

```python
#kmeans로 클러스터링하고 inertia 보기  
inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_2d)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 10), inertias)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
